```python
from sklearn.model_selection import train_test_split

# Features
features = [["RegionID", "SizeRank", "StateName"]]
state_to_number = {}

#encode each state to 0-49, without repeating
def encode_state(state):
    if state not in state_to_number:
        state_to_number[state] = len(state_to_number)
    return state_to_number[state]

home_df['StateName'] = home_df['StateName'].apply(encode_state)

#remove nan values
home_df = home_df.dropna()
print(home_df)
#print(features_cleaned)

# Split the data into features and target
features_cleaned = home_df[["RegionID", "SizeRank", "StateName"]]
target_cleaned = home_df.drop(columns=["RegionID", "SizeRank", "RegionName", "RegionType", "StateName"])

print(features_cleaned)
print(target_cleaned)
```

```python
# split data into training and testing sets using temporal split
# data goes from 1/31/18 to 9/30/24 (monthly) total of 81 months
# training: 1/31/18 to 12/31/22
# testing: 1/31/23 to 9/30/24

# Split the data into training, validation, and testing sets
# Define the split points
train_end = '2022-12-31'
test_start = '2023-01-31'

# Split the data
X_train = features_cleaned.loc[:, features_cleaned.columns]
y_train = target_cleaned.loc[:, :'2022-12-31']

X_test = features_cleaned.loc[:, features_cleaned.columns]
y_test = target_cleaned.loc[:, '2023-01-31':]

"""""
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
"""
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Create a random forest regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf.fit(X_train, y_train)

# Make predictions
train_preds = rf.predict(X_train)
test_preds = rf.predict(X_test)

# Score the model

train_mse = mean_squared_error(y_train, train_preds)
train_r2 = r2_score(y_train, train_preds)

#test_mse = mean_squared_error(y_test, test_preds)
#test_r2 = r2_score(y_test, test_preds)

print(f"Train MSE: {train_mse}")
print(f"Train R^2: {train_r2}")

#print(f"Test MSE: {test_mse}")
#print(f"Test R^2: {test_r2}")
```
